```python
def lwbs(db , value):
    start =0
    end = len(db)-1
    middle = int((end+start)/2)
    
    while not(db[middle] == value):
        if db[middle] > value:
            end = middle -1
        else:
            start = middle +1
        middle = int((end+start)/2)
        print("start:" , start, "end: ", end, "middle: ", middle)
    
    if db[middle]== value:
        print("Found!!")
    else:
        print("Not Found")
    print("start:" , start, "end: ", end, "middle: ", middle)
```

```python
lwbs(db , 59)
```
